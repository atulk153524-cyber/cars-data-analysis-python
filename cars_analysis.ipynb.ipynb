17. Create a bar chart showing the number of cars by Origin.


```python
import matplotlib.pyplot as plt
```

```python
df['Origin'].value_counts().plot(kind='bar')

plt.title('Number of Cars by Origin')
plt.xlabel('Origin')
plt.ylabel('Number of Cars')

plt.show()
```

18. Create a scatter plot between Horsepower and MPG_City.

```python
plt.scatter(df['Horsepower'], df['MPG_City'])

plt.title('Horsepower vs MPG_City')
plt.xlabel('Horsepower')
plt.ylabel('MPG_City')

plt.show()
```

19. Create a histogram of the Weight column.

```python
plt.hist(df['Weight'])

plt.title('Distribution of Car Weights')
plt.xlabel('Weight')
plt.ylabel('Frequency')

plt.show()
```

21. Create a new column called 'Price_Difference' = MSRP - Invoice.


```python
df['Invoice'] = df['Invoice'].astype(str).str.replace('$','').str.replace(',','')
df['Invoice'] = pd.to_numeric(df['Invoice'], errors='coerce')
```

```python
df['Price_Difference'] = df['MSRP'] - df['Invoice']
```
